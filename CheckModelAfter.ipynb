```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
```

```python
dataframe = pd.read_csv('After.csv')
dataframe.head()
```

```python
dataframe.columns
```

```python
clean_data  = dataframe.drop(columns=['Fitur Kontras', 'Fitur Dissimilarity','Fitur Entropi', 'Fitur Homogenitas', 'Fitur Entropi', 'Fitur ASM', 'Fitur Correlations', 'Fitur Energy'])
clean_data = clean_data.replace({
    'Cataract': 0,
    'Myopia': 1,
    'Normal': 2
})
clean_data.head()
```

```python
clean_data.columns
```

```python
len(clean_data.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


x = clean_data.iloc[:,:-1]
y = clean_data.iloc[:,-1:]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

scaler = MinMaxScaler()
x_train_scalled = scaler.fit_transform(x_train)
x_test_scalled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

regressor = LinearRegression()

regressor.fit(x_train_scalled, y_train)

y_pred_continuous = regressor.predict(x_test_scalled)

mse = mean_squared_error(y_test, y_pred_continuous)
print(f"Mean Squared Error: {mse}")
```
